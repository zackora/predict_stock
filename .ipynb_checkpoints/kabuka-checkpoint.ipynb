```python
import datetime
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from pandas_datareader.nasdaq_trader import get_nasdaq_symbols

start = datetime.date(2020, 1, 1)
end = datetime.date.today()
```

```python
#日経平均
df_Nikkei = web.DataReader('NIKKEI225', 'fred', start, end)
df_Nikkei.plot(figsize=(15, 6), fontsize=14)
plt.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=1, fontsize=18)
plt.grid(True)
plt.show()
```

```python
#ダウ平均株価
df_DJI = web.DataReader('^DJI', 'yahoo', start, end)['Adj Close']
df_DJI.plot(figsize=(15, 6),fontsize=14, color='red')
plt.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=1, fontsize=18)
plt.grid(True)
plt.show()
```

```python
df_nintendo = web.DataReader('7974.JP', 'stooq', start, end)['Close']
df_nintendo.plot(figsize=(15, 6), color='red')
df_sony = web.DataReader('6758.JP', 'stooq', start, end)['Close']
df_sony.plot(figsize=(15, 6), color='blue')
plt.show()
```
